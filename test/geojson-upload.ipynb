#### Notebook demonstrating how a geojson file can be uploaded to the TAP API endpoint as part of an HTTP request

```python
import base64
import requests

#endpoint = "https://dw-tap.nrel.gov"
endpoint = "http://localhost:8080"
base_path = "/v1/timeseries/windspeed"

geojson_path = "obstacles-example.geojson"

# with open(geojson_path, "r") as f:
#     geojson = f.read()

# Decoding the contents of the geojson file into Base64
# geojson_bytes = geojson.encode('ascii')
# base64_bytes = base64.b64encode(geojson_bytes)
# base64_geojson = base64_bytes.decode('ascii')

# Example of decoding from Base64 
# base64_bytes = base64_geojson.encode('ascii')
# message_bytes = base64.b64decode(base64_bytes)
# message = message_bytes.decode('ascii')
# print(message)

params = {"height": "50.0m", 
          "lat": 40.712,
          "lon": -74.0059,
          "start_date": "20100302",
          "stop_date": "20100305",
          "vertical_interpolation": "linear",
          "spatial_interpolation": "idw"}

# Form complete url, with parameters
url = endpoint + base_path + "?"
for k, v in params.items():
    url += "%s=%s&" % (k, str(v))
url = url.rstrip('&')
print(url)

files = {'obstacles': open(geojson_path,'rb')}
r = requests.post(url, files=files)

print (r.text)
```
